```python
import tensorflow as tf

w = tf.Variable(2.)

def f(w):
  y = w**2
  z = 2*y + 5
  return z

with tf.GradientTape() as tape:
  z = f(w)

gradients = tape.gradient(z,[w])
print(gradients)

W = tf.Variable(4.0)
b = tf.Variable(1.0)

def hypothesis(x):
  return W*x + b

x_test = [3.5,5,5.5,7]
print(hypothesis(x_test).numpy())

def mse_loss(y_pred,y):
  return tf.reduce_mean(tf.square(y_pred-y))
  
```
